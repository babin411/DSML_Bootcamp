```python
import numpy as np
import pandas as pd
```

```python
columns_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data', sep='\t', names=columns_names)
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles,on='item_id')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['num_of_ratings'] = df.groupby('title')['rating'].count()
```

```python
ratings.head()
```

```python
ratings['num_of_ratings'].hist(bins=70)
```

```python
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='num_of_ratings', data=ratings, alpha=0.5)
```

```python
df.head()
```

```python
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('num_of_ratings', ascending = False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liadliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
similar_to_liarliar = moviemat.corrwith(liadliar_user_ratings)
```

```python
similar_to_liarliar
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending = False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num_of_ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num_of_ratings']>100].sort_values(
'Correlation', ascending = False)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['num_of_ratings'])
```

```python
corr_liarliar[corr_liarliar['num_of_ratings']>100].sort_values(
    by='Correlation',ascending=False
)
```
